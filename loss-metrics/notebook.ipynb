```python
# вычисление интервала для статической значимости 5%
from matplotlib.pylab import norm

mean = 10345
std_dev = 552

x1 = norm.ppf(0.025, mean, std_dev)

x2 = norm.ppf(0.975, mean, std_dev)

print(x1)
print(x2)
```

```python
# вычисление двустороннего p-значения

mean = 10345
std_dev = 552

delta = 11641 - 10345

p1 = norm.cdf(mean - delta, mean, std_dev)
p2 = 1 - norm.cdf(mean + delta, mean, std_dev)

p_value = p1 + p2

print(p_value)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Примерные данные (категории и количество ошибок)
data = {
    'Причина': ['Ошибка A', 'Ошибка B', 'Ошибка C', 'Ошибка D', 'Ошибка E'],
    'Количество': [45, 30, 15, 8, 2]
}

# Создание DataFrame
df = pd.DataFrame(data)

# Сортировка по убыванию
df = df.sort_values(by='Количество', ascending=False)

# Расчёт накопленного процента
df['Накопленный %'] = df['Количество'].cumsum() / df['Количество'].sum() * 100

# Построение столбцов
fig, ax1 = plt.subplots()

ax1.bar(df['Причина'], df['Количество'], color='skyblue')
ax1.set_ylabel('Количество', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Вторая ось для накопленного процента
ax2 = ax1.twinx()
ax2.plot(df['Причина'], df['Накопленный %'], color='red', marker='o', linestyle='-')
ax2.set_ylabel('Накопленный %', color='red')
ax2.tick_params(axis='y', labelcolor='red')
ax2.set_ylim(0, 110)

# Горизонтальная линия на уровне 80% (правило Парето 80/20)
ax2.axhline(80, color='gray', linestyle='dashed')

# Название и сетка
plt.title('Кривая Парето')
plt.grid(True)
plt.show()
```
