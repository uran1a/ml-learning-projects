### **Что такое "статистически значимое отклонение"?**

Когда мы проводим статистический тест (например, t-тест), мы проверяем **гипотезу**. Обычно это нулевая гипотеза (H0) - например, что **разницы между группами нет** или что **среднее равно определенному значению**.

#### **Пример:**

Ты хочешь проверить, отличается ли среднее время реакции у людей после кофеина от 300 мс.
- **H0:** среднее = 300 мс
- **H1 (альтернатива):** среднее != 300 мс
Ты проводишь **t-тест** и получаешь **t-статистику**, которая показывает, насколько твои данные от предполагаемого значения (300 мс).
