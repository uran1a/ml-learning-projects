Если взять от функции шансов натуральный логарифм (по основанию е), то получится функция, которая так и называется - **логарифм шансов**, или **логит-функция**.



Значение логит-функции для 6 часов составляет **0.98**

ШАНСЫ - это соотношение, того, что произойдет (то есть моя команда выиграет) к тому, что не произойдет (то есть моя команда не выиграет).
